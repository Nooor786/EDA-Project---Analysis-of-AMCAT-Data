```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Salary distribution 
sns.histplot(df['Salary'], bins=30, kde=True, color='red', ax=axes[0, 0])
axes[0, 0].tick_params(axis='x')
axes[0, 0].tick_params(axis='y' )
axes[0, 0].set_title('Salary Distribution')
axes[0, 0].set_xlabel('Salary')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: Degree distribution 
sns.countplot(x='Degree', data=df, palette=['pink', 'yellow', 'blue'], ax=axes[0, 1])
axes[0, 1].set_title('Degree Distribution')
axes[0, 1].set_xlabel('Degree')
axes[0, 1].set_ylabel('Count')
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: College GPA distribution 
sns.histplot(df['collegeGPA'], bins=30, kde=True, color='green', ax=axes[1, 0])
axes[1, 0].tick_params(axis='x')
axes[1, 0].tick_params(axis='y')
axes[1, 0].set_title('College GPA Distribution')
axes[1, 0].set_xlabel('GPA')
axes[1, 0].set_ylabel('Frequency')

# Plot 4: Gender distribution 
sns.countplot(x='Gender', data=df, palette=['pink', 'blue'], ax=axes[1, 1])
axes[1, 1].set_title('Gender Distribution')
axes[1, 1].set_xlabel('Gender')
axes[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# First plot: Scatter plot comparing Salary with College GPA 
sns.scatterplot(data=df, x='collegeGPA', y='Salary', color='blue', ax=axes[0])
axes[0].set_title('Relation Between Salary and College GPA', color='blue')
axes[0].set_xlabel('College GPA')
axes[0].set_ylabel('Salary')

# Second plot: Box plot showing Salary distribution across Gender
sns.boxplot(data=df, x='Gender', y='Salary', palette='coolwarm', ax=axes[1])
axes[1].set_title('Salary Distribution by Gender',color='red')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Salary')
plt.tight_layout()
plt.show()
```

### What are the key factors that impact salary levels among job applicants?

```python
# Select only the numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Create a heatmap with only numeric data
plt.figure(figsize=(20, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

### Is there a relationship between educational qualifications, area of specialization, and salary levels?

```python
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(25, 20))  # Increased height from 15 to 20
gs = gridspec.GridSpec(2, 2)

# Box Plot: Salary vs. Degree 
ax1 = fig.add_subplot(gs[0, :])  
sns.boxplot(x='Degree', y='Salary', data=df, palette='viridis', ax=ax1)
ax1.set_title('Comparison of Salary by Degree', fontsize=22)  
ax1.set_xlabel('Degree Type', fontsize=18)
ax1.set_ylabel('Salary Amount', fontsize=18)
ax1.tick_params(axis='x', rotation=45, labelsize=14)  
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Bar Plot: Average Salary by Specialization 
ax3 = fig.add_subplot(gs[1, :])  
sns.barplot(x='Specialization', y='Salary', data=df, errorbar=None, palette='plasma', ax=ax3)  
ax3.set_title('Average Salary by Specialization Type', fontsize=22)  
ax3.set_xlabel('Specialization Area', fontsize=18)
ax3.set_ylabel('Average Salary', fontsize=18)
ax3.tick_params(axis='x', rotation=45, labelsize=14)  
ax3.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### In what ways do individual traits like gender, age, and geographical location influence salary expectations?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot: Average Salary Levels by Gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Salary', data=df, errorbar=None)  
plt.title('Average Salary Levels by Gender', color='orange', fontsize=16)
plt.xlabel('Gender', color='red', fontsize=14)
plt.ylabel('Average Salary',color='red',fontsize=14)
plt.xticks(fontsize=12)  
plt.show()

# Bar plot: Average Salary Levels by JobCity
plt.figure(figsize=(20, 10))
sns.barplot(x='JobCity', y='Salary', data=df, errorbar=None)  
plt.title('Average Salary Levels by JobCity', color='blue', fontsize=20)
plt.xlabel('Job City', color='red', fontsize=20)
plt.ylabel('Average Salary', color='red', fontsize=20)
plt.xticks(rotation=90, fontsize=12)  
plt.show()
```

```python
# Calculate the mean salary for each JobCity
average_salary_per_city = df.groupby('JobCity')['Salary'].mean().reset_index()

# Identify the city with the maximum average salary
city_with_highest_salary = average_salary_per_city.loc[average_salary_per_city['Salary'].idxmax()]

print("City with the highest average salary:")
print(city_with_highest_salary)
```

### Can we detect any trends or patterns in salary distributions among various job titles and locations?

```python
# Salary levels by Designation (bar plot)
plt.figure(figsize=(20, 10))
sns.barplot(x='Salary', y='Designation', data=df, errorbar=None)  
plt.title('Average Salary Levels by Designation', color='red',fontsize=22)
plt.xlabel('Average Salary',color='blue', fontsize=20)
plt.ylabel('Designation', color='blue',fontsize=20)
plt.xticks(rotation=90, fontsize=14)
plt.tight_layout()  
plt.show()
```

```python
# Grouping the data by 'Designation' and calculating the mean salary for each level
avg_salary_by_designation = df.groupby('Designation')['Salary'].mean().reset_index()

# Finding the level with the highest average salary
highest_avg_salary_level = avg_salary_by_designation.loc[avg_salary_by_designation['Salary'].idxmax()]

print("Level with the highest average salary:")
print(highest_avg_salary_level)
```
